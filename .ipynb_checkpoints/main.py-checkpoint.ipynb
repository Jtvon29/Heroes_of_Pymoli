```python
#insert modules
import pandas as pd
import numpy as np
```

```python
#path of the json file
json_path = ("../Heroesofpymoli/Resources/purchase_data.json")
json_pd = pd.read_json(json_path, orient='records')
json_pd.head()
```

```python
#Player Count
Player_count_df = json_pd.groupby('SN')['SN'].nunique()
Player_count_df.count()
```

```python
#Purchasing Analysis (Total)
#Number of unique items
Unique_items = len(json_pd['Item ID'].value_counts())
Unique_items
```

```python
#Average Purchase Price
Avg_purchase_price = json_pd['Price'].mean()
Avg_purchase_price
```

```python
#Total number of purchases
Tot_purchases = json_pd['Item ID'].count()
Tot_purchases
```

```python
#Total revenue
Total_revenue = json_pd['Price'].sum()
Total_revenue
```

```python
#Gender Demographics
#Percentage and count of male players
#Percentage and cout of female players
#Percentage and count of non-gender-specific
Total_gender_count = Player_count_df.count()
Gender_count = json_pd.groupby('Gender')['SN'].nunique()
Gender_percent = (Gender_count/Total_gender_count)*100
Gender_df = pd.DataFrame({"Gender Count": Gender_count, "Gender Percent": Gender_percent})
Gender_df
```

```python
#Purchasing Analysis (Gender)
#Purchase count
Purchase_count = json_pd.groupby('Gender')['Item ID'].count()
#Average purchase price
Avg_purchase_price = json_pd.groupby('Gender')['Price'].mean()
#Total purchase value
Total_purchase_value = json_pd.groupby('Gender')['Price'].sum()
#Normalizd Totals
Normalized_totals = Total_purchase_value/Gender_count
#Dataframe of Purchasing Analysis
Purchasing_df = pd.DataFrame({"Purchase Count":Purchase_count, "Average Purchase Price":Avg_purchase_price,
                              "Total Purchase Value":Total_purchase_value, "Normalized Totals":Normalized_totals})
Purchasing_df
```

```python
#Age Demographics
#Create bins that contain 4 years per bin
bins = [5,10,15,20,25,30,35,40,45]
ranges = ["5-10","11-15","16-20","21-25","26-30","31-35","36-40","41-45"]
```

```python
#Purchase Count separated in bins
json_pd["Age Range"] = pd.cut(json_pd['Age'], bins, labels=ranges)
json_pd.head()
```

```python
#Purchase count
Purchase_count_age = json_pd.groupby('Age Range')['Item ID'].count()
#Average purchase price
Avg_purchase_price_age = json_pd.groupby('Age Range')['Price'].mean()
#Total purchase value
Total_purchase_value_age = json_pd.groupby('Age Range')['Price'].sum()
#Normalizd Totals
Normalized_totals_age = Total_purchase_value_age/Player_count_df.count()
#Dataframe of Purchasing Analysis
Purchasing_age_df = pd.DataFrame({"Purchase Count by Age Range":Purchase_count_age, "Average Purchase Price by Age Range":Avg_purchase_price_age.round(2),
                              "Total Purchase Value by Age Range":Total_purchase_value_age, "Normalized Totals by Age Range":Normalized_totals_age.round(2)})
Purchasing_age_df
```

```python
#Top Spenders
#Identify the top 5 spenders in the game by total purchase value
#SN
#Purchase Count
Purchase_count_top_SN = json_pd.groupby('SN')['Price'].count()
#Avergae Purchase Price
Average_purchase_price_top_SN = json_pd.groupby('SN')['Price'].mean()
#Total Purchase Value
Total_purchase_value_top_SN = json_pd.groupby('SN')['Price'].sum()
#Dataframe of Values
SN_total_purchase_df = pd.DataFrame({"Purchase Count":Purchase_count_top_SN, 
                                     "Average Purchase Price":Average_purchase_price_top_SN.round(2), 
                                     "Total Purchase Value":Total_purchase_value_top_SN})
#Sort by the top 5 spenders
Top_5_spenders_df = SN_total_purchase_df.sort_values("Total Purchase Value", ascending=False)
Top_5_spenders_df.head()
```

```python
#Most Profitable Items
#Identify the 5 most profitable items by total purchase value
#Item ID
#Item Name
#Purchase Count
Purchase_count_top_prof = json_pd.groupby(["Item ID","Item Name"])['Price'].count()
#Item Price
Item_price_top_prof = json_pd.groupby(["Item ID","Item Name"])['Price'].mean()
#Total Purchase Value
Total_purchase_value_top_prof = json_pd.groupby(["Item ID","Item Name"])['Price'].sum()
#Dataframe of values
Most_profitable_df = pd.DataFrame({"Purchase Count":Purchase_count_top_prof,
                                   "Item Price": Item_price_top_prof,
                                   "Total Purchase Value": Total_purchase_value_top_prof})
#Top 5 most profitable items
Top_5_profit_items = Most_profitable_df.sort_values("Total Purchase Value", ascending=False)
Top_5_profit_items.head()
```

```python
#Most Popular Items
#Identify the 5 most popular items by purchase count
#Item ID
#Item Name
#Purchase Count
#Item Price
#Total Purchase Value
Top_5_popular_items = Most_profitable_df.sort_values("Purchase Count", ascending=False)
Top_5_popular_items.head()
```
